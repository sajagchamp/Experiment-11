```python
#dataset creation
import pandas as pd

data={
    "Roll_no.":[100,101,102,101,104],
    "Gender":["M","F","M","F","M"],
    "Department":["Entc","computer","mechanical","computer","computer"],
    "CGPA":[9.03,8.2,7.5,8.2,8.7]
}

df=pd.DataFrame(data)
print("the data set is:\n",df)
print()
#converting data into csv format

df.to_csv("students.csv",index=False) #index =false means that the indexing which is coming would not come here #if index= true than ak column aur banegajisme indexing bhi aayegi

#displays the shape of data set
print("the shape of data set is:",df.shape)
print()
#displying the size of dataset
print("the size of data set is:",df.size)
print()
#displayng informtion about dataset
print(df.info())
print()

#performing statistical analyis on numeric data
#this 25 percent50 percent are quartiles like q1 q2 q3 like q1 ho gaye 25 percent matlab last 25 percent bache than 50 percent ho gya 25 sai upar and 50 perent baccho sai neeche ya equal etc
display(df.describe())
print()
#tells no.of duplicated values
print("no. of duplicated values are:",df.duplicated().sum())
print()
```
